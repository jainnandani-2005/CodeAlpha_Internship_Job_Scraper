## Data Preprocessing

The "£" symbol was removed from the Price column, and the data type was converted from string to float. This preprocessing step was necessary to perform numerical analysis and create visualizations.

```python
df["Price"] = df["Price"].str.replace("£", "", regex=False)
df["Price"] = df["Price"].astype(float)
df.info()
```

Price Distribution (Histogram)

The histogram was created to visualize the distribution of book prices. It shows how book prices are spread across different price ranges. The KDE (Kernel Density Estimation) curve provides a smooth representation of the overall distribution pattern.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price"],bins=8,kde=True)
plt.title("Distribution of Book Prices")
plt.xlabel("Price(£)")
plt.ylabel("Number of Books")
plt.show()
```

Box Plot Analysis

The box plot was created to analyze the spread of book prices. It displays the median, quartiles, minimum and maximum values, and helps identify any potential outliers in the dataset. This visualization provides a clear summary of the price distribution.

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df["Price"])
plt.title("Box Plot of Book Prices")
plt.ylabel("Price (£)")
plt.show()
```

Top 10 Most Expensive Books

This horizontal bar chart displays the top 10 most expensive books in the dataset. It allows for easy comparison of book prices and helps identify the highest-priced books quickly. Sorting the data in descending order makes the visualization more clear and informative.

```python
top_books=df.sort_values(by="Price",ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(data=top_books,x="Price",y="Title")
plt.title("Top 10 Most Expensive Books")
plt.xlabel("Price (£)")
plt.ylabel("Book Title")
plt.show()
```

Scatter Plot Analysis

The scatter plot was created to visualize the distribution of book prices. Each point represents a single book, making it easier to observe the spread of prices and identify any unusual values or patterns in the dataset.

```python
plt.figure(figsize=(10,5))

plt.plot(df["Price"], marker="o")

plt.title("Book Prices")
plt.xlabel("Book Index")
plt.ylabel("Price (£)")

plt.grid(True)
plt.show()
```
